#### **Comparison of spectra for TE: theoretical, simulated, point sources, white noise**

```python
import pandas as pd
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt
```

##### **1. Theoretical and simulated spectra**

```python
%store -r data
```

```python
# We check that the data file columns give Dℓ = ℓ(ℓ+1)Cℓ / 2π in units of μK
data.TE = data.TE/((10**6)**2) # convert to Kelvin
prefactor_Dl = data.L * (data.L + 1) / (2 * np.pi)
cl_TE_teorico = data.TE/prefactor_Dl
data['cl_TE_teorico'] = cl_TE_teorico #add column 
```

```python
# Because we have no data for L=0,1 we fill it with zeros
null_row_0 = pd.DataFrame({'L':0, 'TT':0,'TE':0,'EE':0,'BB':0,'PP':0, 'cl_TE_teorico':0}, index=[0])
null_row_1 = pd.DataFrame({'L':0, 'TT':0,'TE':0,'EE':0,'BB':0,'PP':0, 'cl_TE_teorico':0}, index=[1])
data = pd.concat([null_row_0,data.loc[:]]).reset_index(drop=True)
data = pd.concat([null_row_1,data.loc[:]]).reset_index(drop=True)
```

```python
%store -r nside
%store -r fwhm
```

```python
# Simulated power spectrum
lmax = data.L.iloc[-1]
TE_simulated = hp.synfast(data.cl_TE_teorico, nside = nside, lmax = lmax, fwhm = fwhm) # creates a map from Cl's
cl_TE_simulated = hp.anafast(TE_simulated) # simulated power spectrum from the input simulated TE map
```

```python
%store cl_TE_simulated
```

Important note: the standard way to simulate the power spectra is applying the synfast function to data.cl_TT_teorico, data.cl_EE_teorico, data.cl_BB_teorico and data.cl_TE_teorico together in order to obtain the correct TE power spectrum (the rest will be fine).

##### **4. Plot**

```python
ell = np.arange(len(cl_TE_simulated)) 

plt.figure(figsize=(9, 6))  

plt.plot(ell, TE_cl * ell * (ell + 1) / (2 * np.pi), label='Point sources', linestyle='-', linewidth=2, color = 'deeppink')
plt.plot(ell, cl_TE_simulated * ell * (ell + 1) / (2 * np.pi), label='Simulated CMB', linestyle='-', linewidth=2, color = 'slateblue')
plt.plot(data.L, cl_TE_teorico * data.L * (data.L + 1) / (2 * np.pi), label='Theoretical CMB', linestyle='-', linewidth=2, c = 'midnightblue')
plt.plot(ell, cl_noise_TE_simulated * ell * (ell + 1) / (2 * np.pi), label='Instrumental noise', linestyle='-', linewidth=2, color = 'seagreen')


plt.xlabel('Multipole Moment ($\ell$)', fontsize=14)
plt.ylabel(r'$\ell(\ell+1)C_\ell/2\pi$' + ' [$K^2$]', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig("TE_plot.jpg", dpi = 150)

plt.show()
```
