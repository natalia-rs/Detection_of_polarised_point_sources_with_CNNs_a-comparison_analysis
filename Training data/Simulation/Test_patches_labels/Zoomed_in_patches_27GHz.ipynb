#### **Zoomed-in patches**

```python
import pandas as pd
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt
```

```python
%store -r nside
%store -r ps_map_T
%store -r ps_map_Q
%store -r ps_map_U
%store -r ps_map_P
%store -r map_labels
%store -r r
%store -r indices
```

##### **Define the patch for T, Q, U**

```python
def patch_map(mapp, nside):
    patch_size = np.radians(3)
    
    # Centre of the patch around the maximum
    center_pix_max = np.argmax(mapp)

    theta_max, phi_max = hp.pix2ang(nside, center_pix_max) #rad

    # Centre of the patch around the minimum
    center_pix_min = np.argmin(mapp)

    theta_min, phi_min = hp.pix2ang(nside, center_pix_min) #rad

    fig, (ax1, ax2) = plt.subplots(ncols=2)
    plt.axes(ax1)
    hp.gnomview(mapp, title='Zoomed Map in the max', rot=(np.degrees(phi_max), np.degrees(np.pi/2-theta_max), 0), fig=fig, hold = True, xsize = 50)
    plt.axes(ax2)
    hp.gnomview(mapp, title='Zoomed Map in the min', rot=(np.degrees(phi_min), np.degrees(np.pi/2-theta_min), 0), fig=fig, hold = True, xsize = 50)
    plt.show()

```

##### **T map**

```python
patch_map(ps_map_T, nside)
```

##### **Q map**

```python
patch_map(ps_map_Q, nside)
```

##### **U map**

```python
patch_map(ps_map_U, nside)
```
