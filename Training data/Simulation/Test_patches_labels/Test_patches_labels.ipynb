```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
nside = 1024
npix = hp.nside2npix(nside)
mapa = np.zeros(npix)
fwhm = np.radians(3)
```

```python
P = np.arange(0.01,0.101,0.01) # same len as the PS
r_max = 0.02 #rad
r_min = 0.001
P_min = 0.03
P_max = 0.08
m = (r_max - r_min) / (P_max - P_min)
r = r_min + m * (P - P_min)
r[r < r_min] = r_min
r[r > r_max] = r_max
```

```python
number_ps = len(P)
indices = np.random.randint(0, npix, size=number_ps)
```

```python
# Fill the map with the point sources and their radii
for i, pos in enumerate(indices):       # pos is the pixel index and i is the assignated index from 0-9
    vec = hp.pix2vec(nside, pos)        # converts the pixel index pos to a unit vector on the sphere
    r_i = r[i]
    idx = hp.query_disc(nside, vec, r_i)
    mapa[idx] = 1
   
```

```python
def patch_map_label_maxr():

    # Center of the patch around the maximum radius
    max_value = np.max(r[np.nonzero(r)])
    index_ps_array = np.where(r==max_value)[0][0]
    center_pix_max = indices[index_ps_array]
    theta_max, phi_max = hp.pix2ang(nside, center_pix_max)

    # Center of the patch around the minimum radius
    min_value = np.min(r[np.nonzero(r)])
    index_ps_array = np.where(r==min_value)[0][0]
    center_pix_min = indices[index_ps_array]
    theta_min, phi_min = hp.pix2ang(nside, center_pix_min)

    fig, (ax1, ax2) = plt.subplots(ncols=2)
    plt.axes(ax1)
    hp.gnomview(mapa, title='Zoomed r Max', rot=(np.degrees(phi_max), np.degrees(np.pi/2-theta_max), 0), fig=fig, hold=True, xsize=300)
    plt.axes(ax2)
    hp.gnomview(mapa, title='Zoomed r Min', rot=(np.degrees(phi_min), np.degrees(np.pi/2-theta_min), 0), fig=fig, hold=True, xsize=300)
    plt.show()
```

```python
patch_map_label_maxr()
```

```python
def patch_map_label_maxP():

    # Centre of the patch around the maximum P value
    max_value = P[np.where(r == r.max())][0]
    index_ps_array = np.where(P==max_value)[0][0]
    center_pix_max = indices[index_ps_array]
    theta_max, phi_max = hp.pix2ang(nside, center_pix_max)

    # Centre of the patch around the minimum P value
    nonzero_r = r[r>0]
    min_value = P[np.where(r == nonzero_r.min())][0]
    index_ps_array = np.where(P==min_value)[0][0]
    center_pix_min = indices[index_ps_array]
    theta_min, phi_min = hp.pix2ang(nside, center_pix_min)

    # Plot the zoomed-in maps
    fig, (ax1, ax2) = plt.subplots(ncols=2)
    plt.axes(ax1)
    hp.gnomview(mapa, title=f'Zoomed Map for P = {max_value:.3f}', rot=(np.degrees(phi_max), np.degrees(np.pi/2-theta_max), 0), fig=fig, hold=True, xsize=20)
    plt.axes(ax2)
    hp.gnomview(mapa, title=f'Zoomed Map for P = {min_value:.3f}', rot=(np.degrees(phi_min), np.degrees(np.pi/2-theta_min), 0), fig=fig, hold=True, xsize=20)
    plt.show()

patch_map_label_maxP()
```
