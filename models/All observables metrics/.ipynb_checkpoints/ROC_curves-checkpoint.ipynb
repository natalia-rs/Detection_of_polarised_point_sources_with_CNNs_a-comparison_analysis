```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from scipy.integrate import simps
import pandas as pd
```

```python
plt.figure(figsize=(8, 6))  

thresholds = np.arange(0.1, 1, 0.1)
plt.plot(thresholds, R_list_E, marker='o', label='E', markersize = 4)
plt.plot(thresholds, R_list_EB, marker='o', label='EB', markersize = 4)
plt.plot(thresholds, R_list_B, marker='o', label='B', markersize = 4)
plt.plot(thresholds, R_list_P, marker='o', label='P', markersize = 4)
plt.plot(thresholds, R_list_QU, marker='o', label='QU', markersize = 4)

plt.xlabel(r'Threshold',fontsize=16)
plt.title(r'Reliabilities for the Different Observables Without Foregrounds',fontsize=15, pad=20)
plt.legend(loc='lower right', borderpad=1.2, labelspacing=1.2, fontsize = 12)  
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.subplots_adjust(bottom=0.15, top=0.85)

plt.savefig('Reliability_comparison.jpg', dpi = 200)
plt.show()

```

```python
# Calculate the area behind each curve
area_E = simps(R_list_E, thresholds)
area_EB = simps(R_list_EB, thresholds)
area_B = simps(R_list_B, thresholds)
area_P = simps(R_list_P, thresholds)
area_QU = simps(R_list_QU, thresholds)

# Create a table using pandas
area_reliabilities = {
    'Observable': ['E', 'EB', 'B', 'P', 'QU'],
    'Area': [area_E, area_EB, area_B, area_P, area_QU]
}
area_reliabilities_table = pd.DataFrame(area_reliabilities)
area_reliabilities_table
```
