```python
def sky_ps_map(table_detections, table_spurious, nside=1024, filename='Sky_ps.jpg', title='Sky Point Sources for EB'):
    lon1 = np.array(table_detections['lon(째)'])
    lat1 = np.array(table_detections['lat(째)'])
    indices1 = hp.ang2pix(nside, lon1, lat1, lonlat=True)
    
    if len(table_spurious) > 0:
        lon2 = np.array(table_spurious['lon(째)'])
        lat2 = np.array(table_spurious['lat(째)'])
        indices2 = hp.ang2pix(nside, lon2, lat2, lonlat=True)
    else:
        lon2 = lat2 = []
        indices2 = []

    skymap = np.zeros(hp.nside2npix(nside))
    skymap[indices1] = 1
    skymap[indices2] = 1

    fig = plt.figure(figsize=(10, 6))  

    hp.mollview(skymap, cmap='gray', title=title, cbar=False)
    hp.projscatter(lon1, lat1, lonlat=True, s=1, c='red', marker='o', label='Detections')
    #hp.projscatter(lon2, lat2, lonlat=True, s=1, c='cyan', marker='o', label='Spurious')
    plt.legend()

    plt.savefig(filename, dpi=200)
    plt.show()

    return fig, skymap
```

```python
fig = sky_ps_map(table_detections_E, table_spurious_E, filename='Sky_PS_E_03.jpg', title='Sky Point Sources for E')
```
