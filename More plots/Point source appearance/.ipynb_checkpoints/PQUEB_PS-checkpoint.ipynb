The patch 1439 can be a very representative patch

```python
map_cut_data_QU_obs_squeezed = map_cut_data_QU_obs_squeezed[1439]
```

```python
import matplotlib.pyplot as plt

# Define the cropping coordinates
x_start, x_end = 20, 45
y_start, y_end = 5, 30

# Define the images
images = [
    map_cut_data_P_obs[1439],
    map_cut_data_QU_obs_squeezed[:,:,0],
    map_cut_data_QU_obs_squeezed[:,:,1],
    map_cut_data_E_obs[1439],
    map_cut_data_B_obs[1439]
]

# Define titles for each plot
titles = ["P", "Q", "U", "E", "B"]

# Create a figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 8))  # 2 rows, 3 columns

# Loop through images and titles to create subplots
for i, (ax, image, title) in enumerate(zip(axes.flatten(), images, titles)):
    if i == 0:  # Adjust the first subplot for the bigger P plot
        im = ax.imshow(image[y_start:y_end, x_start:x_end], cmap='viridis')  # Use the desired colormap
    else:
        im = ax.imshow(image[y_start:y_end, x_start:x_end], cmap='viridis')
    ax.set_title(title, fontsize=17)  # Set mini titles
    ax.axis('off')  # Turn off axis
    # Add colorbar for each plot
    cbar = fig.colorbar(im, ax=ax, pad=0.02, fraction=0.046)
    cbar.ax.tick_params(labelsize=12)  # Adjust colorbar label size
    cbar.set_label('Kelvin', rotation=90, labelpad=7, fontsize=12)  # Add units to color scale

# Remove the empty subplot
fig.delaxes(axes[1, 2])

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.2, hspace=0.4)

# Show the composite plot
plt.savefig("comparisonpointsource.jpg", dpi=300)
plt.show()
```
